# Movies watched by children

```python
child = df[df["Type"].str.contains("Kids' TV")]
child.head(3)
```

```python
child.shape
```

```python
count_child = len(child)
count_others = (len(df['Type']))-count_child 
```

```python
# Data to plot
y = np.array([count_child, count_others])
mylabels = ["Children", "Others"]
colors = ['skyblue','gold']
myexplode = [0.3, 0]

def absolute_value(val):
    a  = np.around(val/100.*y.sum(),0)
    return a
plt.figure(figsize=(8,8))
plt.pie(y, labels = mylabels, explode = myexplode,autopct=absolute_value,colors = colors)
plt.legend()
plt.show() 
```

# Total number of Comedies 

```python
# Selecting only the Standup Comedies from listed_in
stand_comedies = df[df['Type'].str.contains('Comedies')]
stand_comedies.head(5)
```

```python
count_stand_comedies = len(stand_comedies)
```

```python
count_stand_comedies
```

```python
count_remain = (len(df['Type']))-count_stand_comedies 
```

```python
# Data to be plotted
y = np.array([count_stand_comedies, count_remain])
labels = ["Standup Comedies", "Others"]
pallet = ['green','gray']
my_explode = [0.1, 0]

def absolute_value(val):
    a  = np.around(val/100.*y.sum(),0)
    return a
plt.figure(figsize=(8,8))
plt.pie(y, labels = labels, explode = my_explode,autopct=absolute_value,colors = pallet)
plt.legend()
plt.show()
```

We can clearly see the number of Standup Comedies movies with the help of Data Visualization
