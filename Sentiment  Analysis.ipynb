# Sentimental Analysis of presidential elections 

```python
lm = WordNetLemmatizer()
```

```python
# This is data preprocessing
def clean(text):
    text = str(text).lower()
    text = re.sub('[^a-z]',' ',str(text))
    text = text.split()
    text = [ lm.lemmatize(j) for j in text if j not in set(stopwords.words('english'))]
    text = ' '.join(str(k) for k in text)
    return text
```

```python
# create function to get polarity and subjectivity
# subjectivity = if the sentence is subjective or objective - subjective sentence refers to personal opinions, emotions or people's judgemnet whereas objective sentence refers to factual informations. It ranges from [0,1].

# Polarity -- Float value that lies between [-1,1], where 1 means positive sentiment and -1 means -ve sentiment.
```

# Donald trump tweets sentiment Analysis

```python
trump_tweets = data_mixed.query('(Candidate == "Trump")').sort_values('user_followers_count',ascending= False).drop_duplicates(['user_name'])[['tweet','country']]
trump_tweets = trump_tweets.dropna().loc[trump_tweets.country =='US']
trump_tweets.reset_index(inplace = True, drop=True)
```

```python
trump_tweets
```

```python
trump_tweets['cleantext'] = trump_tweets['tweet'].apply(clean)
```

```python
trump_tweets
```

```python
def getsubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
def getpolarity(text):
    return TextBlob(text).sentiment.polarity
def getAnalysis(score):
    if score<0:
        return 'negative'
    elif score ==0:
        return 'neutral'
    else:
        return 'positive'
```

```python
trump_tweets['subjectivity'] = trump_tweets['cleantext'].apply(getsubjectivity)
```

```python
trump_tweets['polarity'] = trump_tweets['cleantext'].apply(getpolarity)
```

```python
trump_tweets['analysis'] = trump_tweets['polarity'].apply(getAnalysis)
```

```python
trump_tweets.head()
```

```python
trump_tweets.analysis.value_counts() # to check the subjectivity
```

```python
trump_tweets.analysis.value_counts(normalize=True)*100
```

```python
#plot the graph for analysis tweets
plt.figure(figsize=(10,6))
(trump_tweets.analysis.value_counts(normalize=True)*100).plot.bar()
plt.ylabel('%age of tweets')
```

```python
from wordcloud import WordCloud, STOPWORDS
def word_cloud(wd_list):
    stopwords = set(STOPWORDS)
    all_words = ' '.join([text for text in wd_list])
    wordcloud = WordCloud(background_color = 'white',stopwords = stopwords,
                         width = 1600,height = 800,random_state=2,colormap = 'jet',max_words = 100,max_font_size=250).generate(all_words)
    plt.figure(figsize=(12,10))
    plt.axis('off')
    plt.imshow(wordcloud)
```

```python
word_cloud(trump['tweet'][:5000])
```
