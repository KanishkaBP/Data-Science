```python
#To convert string into 0's and 1's
from sklearn import preprocessing
lb = preprocessing.LabelEncoder()
df_sort['Sex'] = lb.fit_transform(df_sort['Sex'])
```

```python
# Simplifying the Age column
from sklearn import preprocessing
lb = preprocessing.LabelEncoder()
df_sort['Age'] = lb.fit_transform(df_sort['Age'])
```

```python
# Spliting the dataset into train dataset and test dataset
from sklearn.model_selection import train_test_split
x = df_sort[['Pclass','Sex','Age']]
y = df_sort[['Survived']]
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
x
```

```python
y
```

```python
# Using Naive Bayes Library to get the accuracy
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
# Predicting our model
y_pred = gnb.predict(x_test)
```

```python
# Getting the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
# Getting the accuracy in %age
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test,normalize=True)*100
```
